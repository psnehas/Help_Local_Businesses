# Covid 19 Questionaire Responses' Analysis

### Importing Essential LIbraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyplot
%matplotlib inline
```

### Importing The dataset


```python
responses_df = pd.read_csv("responses.csv")
responses_df.head()
```

```python
# Questions asked 

Q1 = "Do you usually get your books from Amazon/Barnes & Noble/ other big retailers?"
Q2 = "Why do you shop at a larger retail store rather than shopping at a local bookstore?"
Q3 = "Amazon's model allows it to make books a little bit cheaper. Would you be willing to pay 10% more than what you would pay for a book at a larger retailer if it means helping the local business stay afloat? "
Q4 = "What would motivate you to buy books from your local bookstore?"
```

```python
# Since the Timestamp is not relevant for this anlysis - we are dropping the feature

responses_df.drop(['Timestamp'], axis = 1, inplace=True)
responses_df.head()
```

## DO PEOPLE BUY BOOKS FROM AMAZON?

```python
# Data based on the resposes received for question 1

Q1_no = (responses_df[Q1]=="No").sum()/len(responses_df)
Q1_yes = (len(responses_df) - (responses_df[Q1]=="No").sum())/ len(responses_df)

print("Based on the reponse received, {:.2f} % responded they buy books from big retailers and {:.2f} % responded they dont.".format(Q1_yes*100, Q1_no*100))
```

```python
# VISUAL PRESENTATION FOR QUESTION 1 RESPONSES


labels = ['Yes', 'No']
responses = [Q1_yes, Q1_no]
ex = (0, 0.3)  

fig1, ax1 = plt.subplots()
ax1.pie(responses, explode=ex, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title(Q1)
plt.show()
```

```python
Q1_reasons = ['convenience','cheaper', 'convenience', 'variety','convenience', 'cheaper', 'convenience','variety','variety','convenience','cheaper','variety', 'variety','convenience','convenience','convenience','variety','convenience','convenience','convenience']
```

```python
# Response counts based on why people buy at large book retail stores

conv = Q1_reasons.count('convenience')
var = Q1_reasons.count('variety')
chp = Q1_reasons.count('cheaper')
l = len(Q1_reasons)


print("Based on the reasons received on why people buy from large book retailers, {:.2f} % responded convenience, {:.2f} % responded variety, and {:.2f} % responded it is cheaper at bigger retail stores.".format(conv*100/l, var*100/l, chp*100/l))
```

```python
# VISUAL PRESENTATION FOR QUESTION 1 RESPONSES


labels = ['convenient', 'variety', 'cheaper']
reasons = [conv, var, chp]
ex = (0, 0.3, 0.2)  

fig1, ax1 = plt.subplots()
ax1.pie(reasons, explode=ex, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title("Reasons for buying books from big retail stores")
plt.show()
```

### WHAT WOULD PEOPLE MOTIVATE FROM LOCAL STORES

```python
motivations = ['support', 'services', 'support', 'services', 'support', 'support', 'price', 'services', 'services', 'variety', 'support', 'services', 'services', 'services', 'price', 'support', 'price', 'services', 'support', 'services', 'variety', 'support', 'varity']
```

```python
# Response counts based on why people would be interested to buy books from local stores

supp= motivations.count('support')
serv = motivations.count('services')
var = motivations.count('variety')
prc = motivations.count('price')
m = len(motivations)


print("Based on the motivations received on why people would buy books from local stores, {:.2f} % would like to support, {:.2f} % would like services inside the local book stores, {:.2f} % responded they would like variety, and {:.2f} % responded they would like cheaper price in local book stores.".format(supp*100/m, serv*100/m, var*100/m, prc*100/m))
```
